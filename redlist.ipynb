```python
import glob, json, os, subprocess
```

```python
def subprocess_check(*args, **kwargs):
    if len(args) == 1 and type(args[0]) == str:
        kwargs['shell'] = True
    p = subprocess.Popen(*args, stdout=subprocess.PIPE, stderr=subprocess.PIPE, **kwargs)
    (out, err) = [x.strip() for x in p.communicate()]
    ret = p.wait()
    if not ret:
        return (err + '\n' + out) if (err and out) else (err + out)
    raise Exception(
        'Call to subprocess_check failed with return code %d\nStandard error: "%s"\nStandard out: "%s"' % 
        (ret, err, out))
```

```python
for zipfile in glob.glob('redlist/*.zip'):
    dest = os.path.splitext(zipfile)[0]
    if os.path.exists(dest):
        print '{dest} already exists, skipping'.format(**locals())
    else:
        subprocess_check(['mkdir', dest + '.tmp'])
        subprocess_check(['unzip', zipfile, '-d', dest + '.tmp'])
        os.rename(dest + '.tmp', dest)
        print 'Unzipped {zipfile} into {dest}'.format(**locals())
    if dest == 'redlist/CORALS':
        continue
    geojson = dest + '.geojson'
    if os.path.exists(geojson):
        print '{geojson} already exists, skipping'.format(**locals())
    else:
        shpfile = dest + '/' + os.path.basename(dest) + '.shp'
        subprocess_check(['ogr2ogr', '-f', 'GeoJSON', geojson + '.tmp.geojson', shpfile])
        os.rename(geojson + '.tmp.geojson', geojson)
        print 'Converted {shpfile} to {geojson}'.format(**locals())
```
